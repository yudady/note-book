```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
# openCV的cv2.imread()导入图片时是BGR通道顺序
image = cv2.imread("angel.jpg")

#提取R、G、B分量
b,g,r = cv2.split(image)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
plt.imshow(r)
```

```python
img1 = cv2.imread("angel.jpg",1)
plt.imshow(img1)
plt.show()
```
